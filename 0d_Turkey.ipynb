```python
import geopandas as gpd
```

```python
df_turkey = gpd.read_file('https://raw.githubusercontent.com/alpers/Turkey-Maps-GeoJSON/master/tr-cities.json')
df_turkey.geometry = df_turkey.buffer(1e-2).simplify(1e-3)
df_turkey = df_turkey.dissolve()
df_turkey.plot()
```

# Serialize

```python
df_turkey.to_file('AOIs/turkey.geojson', driver='GeoJSON')
```
